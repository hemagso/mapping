```python
import geopandas as gpd
import random
import csv
import shapely.geometry

random.seed(42)
```

```python
df = gpd.read_file("./data/income_map")
df.head(50)
```

```python
def generate_coordinates(geometry,n):
    "Generate random points inside a geometric shape"
    ret = []
    x_min, y_min, x_max, y_max = geometry.bounds
    while len(ret) < n:
        pt_x = x_min+random.uniform(0,1)*(x_max-x_min)
        pt_y = y_min+random.uniform(0,1)*(y_max-y_min)
        pt = shapely.geometry.Point(pt_x,pt_y)
        if geometry.contains(pt):
            ret.append(pt)        
    return ret
```

```python
df.sort_values("COD_SETOR",inplace=True)

results = []

for index, row in df.loc[:,["COD_SETOR", "geometry"]].iterrows():
    clist = generate_coordinates(row["geometry"],10)
    #Trocamos lat e long
    records = [(row["COD_SETOR"], idx, point.y, point.x) for idx, point in enumerate(clist)]
    results = results + records
    
len(results)
    
```

```python
with open("./data/coordinates.csv","w") as csvfile:
    writer = csv.writer(csvfile,lineterminator="\n")
    writer.writerows(results)
```
